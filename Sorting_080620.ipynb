#### Bubble Sort

```python
from IPython.display import Image
```

```python
PATH = "C:/Algorithms_DataStructures/"
Image(filename = PATH + "Bubble Sort.png", width=400, height=400)
```

```python
def bubble_sort(a_list):
    for pass_num in range(len(a_list) - 1, 0, -1):
        for i in range(pass_num):
            if a_list[i]>a_list[i+1]:
                a_list[i+1],a_list[i]=a_list[i],a_list[i+1]        
        
    return a_list

def bubble_sort1(a_list):
    for i in range(len(a_list),0,-1):
        for ii in range(i-1,0,-1):
            if a_list[ii]<a_list[ii-1]:
                a_list[ii-1],a_list[ii]=a_list[ii],a_list[ii-1]
                
    return a_list
                
        
        
    
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print(bubble_sort(a_list))
print(bubble_sort1(a_list))
```

#### Selection Sort

```python

PATH = "C:/Algorithms_DataStructures/"
Image(filename = PATH + "selection_sort.jpg", width=500, height=500)
```

```python
def selection_sort(a_list):
    for i in range(len(a_list)-1,-1,-1):
        pos = i
        for j in range(i+1):
            if a_list[j]>a_list[pos]:
                a_list[pos],a_list[j]=a_list[j],a_list[pos]
    return a_list            
            
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print(selection_sort(a_list))
```

##### Insertion Sort

```python
PATH = "C:/Algorithms_DataStructures/"
Image(filename = PATH + "Insertion Sort.jpg", width=500, height=500)
```

```python
def insertion_sort(a_list):
    for i in range(len(a_list)):
        k=a_list[i]
        j=i-1
        while ((k<a_list[j]) & (j>=0)):
            a_list[j+1]=a_list[j]
            j=j-1
        a_list[j+1]=k
    return a_list
                        

a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print(insertion_sort(a_list))
```

##### Shell Sort

```python
PATH = "C:/Algorithms_DataStructures/"
Image(filename = PATH + "Shell Sort.png", width=500, height=500)
```

```python
def insertion_sort_gap(a_list,g):
    for i in range(g,len(a_list),g):
        k=a_list[i]
        j=i-g
        while ((k<a_list[j]) & (j>=0)):
            a_list[j+g]=a_list[j]
            j=j-g
        a_list[j+g]=k
    return a_list

def shell_sort(a_list):
    sublist = len(a_list)
    while (sublist>0):
        for i in range(sublist):
            insertion_sort_gap(a_list,sublist)
        sublist = sublist//2
        
    return a_list    
        
        
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
gap=3
print(insertion_sort_gap(a_list,gap))
print(shell_sort(a_list))
```

```python
def merge_sort(a_list):
    if len(a_list)>1:
        mid = len(a_list)//2
        left = a_list[0:mid]
        right = a_list[mid:len(a_list)]
        
        merge_sort(left)
        merge_sort(right)
        
        i=0
        j=0
        k=0
        
        
        while (i<len(left) and j<len(right)):
            if left[i]<right[j]:
                a_list[k]=left[i]
                i=i+1
            else:
                a_list[k]=right[j]
                j=j+1
            k=k+1
            
            
        while (i<len(left)):
            a_list[k]=left[i]
            i=i+1
            k=k+1
            
            
        while (j<len(right)):
            a_list[k]=right[j]
            j=j+1
            k=k+1
                
        
    return a_list
        
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
merge_sort(a_list)
    
```
